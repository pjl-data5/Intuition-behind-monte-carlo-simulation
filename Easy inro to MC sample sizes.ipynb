# Monte Carlo simulation

Say one has a random variable and you're unsure of its true probability distribution. By making use of sample techniques such as MC simulations, might yield a solution. Or at least, point you into the right direction. Conventionally, plotting a histogram visualises a distribution. Thus here we are drawing from a simulated normal distribution and plotting these samples. This is done over varying sample sizes to showcase the efficacy of the law of large numbers for choice sample sizes for MC simulations.

```python
# Import relevant packages
from numpy.random import normal
import matplotlib.pyplot as plt
# Renders plots in Jupyter
%matplotlib inline

# Define the normal distribution parameters (arbitrary)
mu = 150
sigma = 15

# Generate MC smaples of varying sizes
sizes = [10, 50, 100, 1000]

for i in range(len(sizes)):
    sample = normal(mu, sigma, sizes[i])
    
    # Render plots
    plt.figure(figsize=(8, 8))
    plt.subplot(2,2,i+1) # so there's only one hist function
    plt.hist(sample, bins=20, ec="black", color="#f54254" )
    plt.title('%d samples' % sizes[i]) # format
    plt.show()
    
```

Visually it is apparent that larger sample sizes yield more accurate estimations.
